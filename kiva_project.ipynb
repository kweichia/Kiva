```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

## 導入資料

```python
df = pd.read_csv('kiva_data.csv')
df.head()
```

## 檢查資料

資料集中的每一橫列代表 Kiva 提供給特定項目的一筆貸款。 `loan_amount` 表示項目的金額（美元）、 `activity` 為項目的類型、 `country` 是項目所在的國家， `gender` 則是貸款申請人的性別。

```python
df.head(25)
```

## 長條圖 Bar Charts 

### 使用長條圖先觀察按國家Kive貸款項目的平均規模

```python
f, ax = plt.subplots(figsize=(15, 10))
ax = sns.barplot(x=df.country, y=df.loan_amount)
```

```python
import matplotlib.ticker as mtick

# Creates the figure
f, ax = plt.subplots(figsize=(15, 10))

# Plot the data
sns.barplot(data=df, x="country", y = "loan_amount")

# Use part of the code above to format the y-axis ticks below this line
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
```

### 加入性別變量觀察

```python
f, ax = plt.subplots(figsize=(15, 10))

sns.barplot(data=df, x="country", y="loan_amount", hue='gender')

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
plt.show()
```

```python
sns.set_palette('Accent')

sns.set_style('whitegrid')

plt.figure(figsize=(25, 15))

ax.set_title('Average Loan Amounts Given by Kiva, by country, by gender (in $ USD)')

ax = sns.barplot(data=df, x='country', y='loan_amount', hue='gender')
plt.show()
```

平均來說，男性申請者從Kiva那邊獲得更多的貸款。

薩爾瓦多在貸款金額方面性別差異最小，而柬埔寨跟巴基斯坦的差異最大。

## 箱形圖 Box Plots

### 使用箱形圖來比較各個國家的貸款金額分佈

```python
plt.figure(figsize=(16, 10))
sns.set_palette('Pastel1')
sns.boxplot(data=df, x='country', y='loan_amount')
plt.show()
```

肯亞的貸款金額的分佈範圍最廣，而柬埔寨較有可能收到大筆的貸款金額。

### 使用箱形圖來比較項目的類型貸款金額分佈

```python
plt.figure(figsize=(16, 10))
sns.boxplot(data=df, x='activity', y='loan_amount')
plt.show()
```

農業類型`Farming`的項目收到的貸款金額較為大筆。 

## 小提琴圖 Violin Plots

### 使用小提琴圖來比較項目類型的貸款金額分佈

```python
plt.figure(figsize=(16, 10))

sns.violinplot(data=df, x="activity", y="loan_amount")
plt.show()
plt.savefig('loan_by_activity')
```

### 使用小提琴圖來觀察各個國家的貸款金額分佈

```python
plt.figure(figsize=(16, 10))
sns.violinplot(data=df, x='country', y='loan_amount')
plt.show()
plt.savefig('loan_by_country')
```

## 分邊小提琴圖 Split Violin Plots

### 加入性別變量作分組

```python
sns.set_palette("Spectral")
plt.figure(figsize=(18, 12))
sns.violinplot(data=df, x='country', y='loan_amount', hue='gender', split=True)
plt.show()
plt.savefig('loan_by_country&gender')
```

巴基斯坦和柬埔寨，女性平均獲得的貸款金額低於男性。

在薩爾瓦多，性別貸款更加平等。

### 使用小提琴圖來比較項目類型的貸款金額分佈，加入性別分組

```python
sns.set_palette("Spectral")
plt.figure(figsize=(18, 12))
sns.violinplot(data=df, x='activity', y='loan_amount', hue='gender', split=True)
plt.show()
plt.savefig('loan_by_activity&gender')
```

## 分析

平均來說，男性申請者能夠從Kiva那邊獲得金額更多的貸款。


薩爾瓦多在貸款金額方面性別差異最小，相對更符合性別平等。
柬埔寨跟巴基斯坦的差異最大，女性平均獲得的貸款金額較男性為低。


根據這些資料，可以建議Kiva盡量縮小貸款之間的性別差距。

- 給予女性申請者更多的曝光機會 
- 使借款金額盡量相等

或許都能夠弭平性別在貸款金額上的不平等，緩和貧窮女性的雙重剝削情況。
